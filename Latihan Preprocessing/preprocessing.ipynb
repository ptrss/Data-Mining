```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read the dataset
dataset = pd.read_csv('Data.csv')
```

```python
# Memilih fitur dan target yang akan digunakan
# ===============================
x = dataset[['Country', 'Age', 'Salary']].values

# Memilih label
y = dataset['Purchased'].values
```

```python
print(x)
```

```python
print(y)
```

```python
# Mengisi atribut yang kosong menggunakan nilai rata-rata(mean) dengan SimpleImputer
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(x[:, 1:3])
x[:, 1:3] = imputer.transform(x[:, 1:3])
```

```python
# Cek kembali value setelah atribut yang sebelumnya kosong diberikan nilai rata-rata
print(x)
```

```python
# Mengubah kolom "Country" ke dalam bentuk matriks menggunakan ColumnTransfer dan OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

```python
# Cek kolom "Country" setelah diubah ke dalam bentuk matriks
print(x)

# Setelah dataset dimodifikasi, teradapat 30 baris data
```

```python
# Mengubah value label "Purchase" ke dalam bentuk matriks menggunakan LabelEncoder
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(y)
```

```python
# Cek value label setelah diubah ke dalam bentuk matriks
print(y)
```

```python
# Melakukan train test split dengan test_size sebesar 20%
from sklearn.model_selection import train_test_split

# Split 24:6
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
# Cek banyak data yang digunakan untuk training
print(x_train)
```

```python
# Cek banyak data yang digunakan untuk testing
print(x_test)
```

```python
# Cek banyak label yang digunakan untuk training
print(y_train)
```

```python
# Cek banyak label yang digunakan untuk testing
print(y_test)
```

```python
# Scaling
from sklearn.preprocessing import StandardScaler

# Hanya diterapkan kepada x_train dan x_test
sc = StandardScaler()
x_train[:, 3:] = sc.fit_transform(x_train[:, 3:])
x_test[:, 3:] = sc.transform(x_test[:, 3:])
```
